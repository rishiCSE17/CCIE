# XML and XPATH 
* Both on-box and off-box tools uses XML to interact with Junos infra. 
* Request-Response model, Extensive markup language 
* Tag based : `<tag> data </tag>`
* Shorthand for tags without any data : `<tag\>` = `<tag></tag>`
* Any hierarchy can be represented
```
<course>
    <name> Intro to python </name>
    <sections>
        <sec1> data </sec1>
        <sec2> data </sec2>
        <sec3> data </sec3>
    </sections>
</course>
```
* Namespace: `<tag xmlns=NAMESPACE>` its possible to have multiple namespace in a single XML document. namespace is provided as a URL using __xmlns__ attribute.

## XML Scehema Definition (XSD)
An XML doc represent a hierarchical structure, There are many way to express a single struct thus we need a standardization process called XML schema def or __XSD__. 
* defined the elements allowed in a XML doc and its hierarchy 
* all XML users will use same tags 
* XSD is written in an XML format 

The junos CLI uses XML API to communicate with the junos infra, every commnad is translated to XML docs. use `CMD | display xml rpc` command to see the equivalant XML RPC request. and use `CMD | display xml` command to see the XML resposne 
```
> show route active-path | disp xml rpc

<get-route-information>
    <active-path/>
</get-route-information>
```

Junos uses XSD specified for each resease, there is seperarte schemas for operational and configuration data, XSD is specified in namespace of the openning tag.  

## Some more terms 
* Element nodes : `<tag attribute1=value1> Text_node </tag>`. Elemet nodes can be nested.
* Document node : root of the XML doc, root element is a child of document node
* __XPATH__ Axes
    * _Ansestor_ : from parent to root
    * _Desendent_ : from child to leaf
    * _self_ : me
    * _sibling_ : having common parent 
    * _Child_ : One level down
    * _Parant_ : One level up
* Operator 
    * _Logical_ : `AND,OR`
    * _Comparison_ : `=, !=, <, >`
    * _numarical_ : `+, -, *`
    
## XPATH 
* Select specific XML node using text notation, similar to UNIX file path.  
* e.g. `/config/system/root-authentication/encrypted-password/`
* XPATH can be __absolute__ or __relative__0 

## Predicates 
mayches a specific parameter within an XML data using `/XPATH[Condition]`

# XML-API & NETCONF

## Why NETCONF
The client eshtablishes a TCP port830 to the router, SSH params are negotiated and client starts sending config messages in XML format.  

* Orchestrating a number of network devices remotely using automation tool securely
* RFC 6241
* runs on TCP port 830 by default
* a successor of Junos-Script API 
* __NETCONF Protocl Layers__

| Layer | Task |
|---    | ---  |
|Content|  __RPC req__, resp __paylaod__ and some additional variables   |
|Operation|`<lock>,<unlock>`: __ConfigDB__, <br> `<get>`,`<get-config>`,`<edit-config>`,`<copy-config>`,`<delete-config>`:__Config__, <br> `<commit>,<discard-changes>,<validate>`:__Config change__, <br> `<create-subscription>,<close-session>,<kill-session>`: __Session info__     |
|Message|`<rpc>`:__Req__, `<rpc-reply>`:__ACK__, `<notification>`:one way Server to client on __subscription ack__ |
|Transport|__SSH__(Junos uses),TLS,HTTP(S)|

```
with Device(host='172.25.11.1',user='lab', password='lab123) as dev:
    
    #use the RPC name replacing '-' with '_' to get the method name
    route_xml_elem = dev.rpc.get_route_information(table='inet.0')
    
    #fetch everything under <rt> element
    list_of_routes = route_xml_elem.findall('.//rt')
    
    for routes in list_of_routes:
        r=route.findtext('rt-destination').strip()
        p=route.findtext('et-entry/protocol-name).strip()
        
        print(f'Route {r} Protocol {p}')
```

To get XML output

```

dev = Device(host = 'vMX-1', user='lab', password='lab123')
conf = Config(dev)
```
Then follow the steps
* lock the configuration using `lock()` 
* load the new config using `load()`
* In case of rollback use `rollback()` or rescue to a config using `rescue()`
* commit the config using `commit()`
* unlock the config `unlock()`

Another example of using `config` object under context manager using `set` command

```
from jnpr.junos import Devices
from jnpr.junos.utils.config import Config 

data = 'set system services netconf traceoptions files trest.log'

#auto open() and close()
with Device(host = '172.25.11.1', user='lab', password='lab123') as dev:
    
    #auto lock() and unlock()
    with Config(dev, mode='exclusive') as conf: 
    
        conf.load(data,format='set')
        
        #config difference 
        diff = conf.diff()  
        
        if diff == None:
            print('Config is up to date')
        else:
            print(f'Config update... \n{diff}')
            conf.commit()
```

## Exception handling in PyEZ
* Exceptions are defined in `jnpr.junos.exception` module. 
* Put more specific exception at top 

The generic code layout would look like

```

def  get_bgp_state(host, uname, pass):
    from jnpr.junos import Device 
 
    peer_est = 0  #num of established peer
    peer_oth = 0  #num of other tha eshtablished peer

    with Device(host=host, user=uname, password=pass) as dev:
        bgp_data   = dev.rpc.get_bgp_summary_information()   # RPC call
        nei_states = bgp_data.xpath('bgp_peer/peer_state')   # relative XPATH
        
        for state in nei_states:
            if state.text == 'Eshtablished':
                peer_est += 1
            else:
                peer_oth += 1
                
    if ceil(peer_oth / (peer_est + peer_oth)):
        exit(1)
        
    return 0
```

```
request security pki generate-key-pair \
                  certificate-id REST_CERT

request security pki local-certificate \
                  generate-self-signed \
                  certificate-id REST_CERT \
                  subject CN = MY-TEST \
                  domain-name vMX-1.example.com \
                  email lab@example.com
configure 
    set system services rest https server-certificate REST_CERT
    set system services rest https port 4040  # default port 3443 
```
### Access control in REST
`set rest control [option]`

| Option | Meaning |
| --- | --- |
| `alloed-source` | list of allow |
| `apply-goup`    | group from which inherit configuration data |
| `apply-group-exept` | don't inherit confing data from these groups |
| `connection-limit [1...1024]` | maximum number of simultainous connections |

## Querying through REST API

`[proto]://[host]:[port]/rpc/[procedure]` i.e. `http://172.25.11.1:3000/rpc/get-interface-information` Junos by default responses using XML format. 

* _protocol_ : HTTP, HTTPS
* _host_ : IP address or name 
* _port_ : TCP port number 
* _procedure_ : Junos only supports __`GET`__ and __`POST`__ methods 

to send a request using `curl` : `curl -u "UN:PW" proto://host:port/rpc/proc @format=[text|json]?[param1=val1]&[param2=val2]`. The `format` variable selects the output type for additional parameter such as fixing an interface name use `params` starting with `?` and seperated by `&`. 

For `POST` use `-d` in `curl`, it can carry multiple REST requests with one RPC call. the Requests are put into a payload which is sent over the `POST` method. A single HTTP response will carry all the responses seperated by a special seperator. 

## Junos REST API Explorer
A GUI tool to interact with Junos via REST. Send single or multiple RPC, choose methods, output format and provide RPC call with login credentials. The __`Stop on Error`__ option if checked the processing will stop if any RPC call fails, otherwise it will keep executing all except the one that fails. For __Multiple RPC__ seperate the requests by empty line. 
* Enable it by `set system services rest enable-explorer` then `commit`
* open it a browser with `http://[router-ip]:3000`
